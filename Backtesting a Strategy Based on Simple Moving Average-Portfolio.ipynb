## Backtesting a Strategy Based on Simple Moving Average

```python
# Descargar datos de MSFT y guardarlos en un archivo CSV
ticker = "MSFT"
start_date = "2023-01-01"
end_date = "2023-12-31"
msft = yf.Ticker(ticker)
data = msft.history(start=start_date, end=end_date)
csv_file_path = 

data.to_csv(csv_file_path)


# Definición de la clase SmaSignal para el backtest
class SmaSignal(bt.Signal):
    params = (('period', 20), )
    
    def __init__(self):
        self.lines.signal = self.data - bt.ind.SMA(period=self.p.period)

# Cargar datos desde el archivo CSV local para el backtest
data = bt.feeds.GenericCSVData(
    dataname=csv_file_path,
    fromdate=datetime(2023, 1, 1),  
    todate=datetime(2023, 12, 31),
    nullvalue=0.0,
    dtformat=('%Y-%m-%d %H:%M:%S%z'),  
    datetime=0,
    open=1,
    high=2,
    low=3,
    close=4,
    volume=5,
    openinterest=-1
)

# Configuración y ejecución del backtest
cerebro = bt.Cerebro(stdstats = False)
cerebro.adddata(data)
cerebro.broker.setcash(1000.0)
cerebro.add_signal(bt.SIGNAL_LONG, SmaSignal)
cerebro.addobserver(bt.observers.BuySell)
cerebro.addobserver(bt.observers.Value)

print(f'Starting Portfolio Value: {cerebro.broker.getvalue():.2f}')
cerebro.run()
print(f'Final Portfolio Value: {cerebro.broker.getvalue():.2f}')
```
